```python
import pandas as pd

# Load JSON
data = pd.read_json('delhi_metro_trips.json')

data
```

```python
# Removing extra spaces in station names
# Standardizing station naming conventions

data['From_Station'] = data['From_Station'].str.strip()
data['To_Station'] = data['To_Station'].str.strip()

data
```

```python
# Handling missing values in ticket types

data['Ticket_Type'] = data['Ticket_Type'].fillna('Unknown')
data['Remarks'] = data['Remarks'].fillna('No Remarks')

data
```

```python
# Converting date columns into datetime format

data['Date'] = pd.to_datetime(data['Date'])

data
```

```python
# Ensuring numeric columns are correctly formatted

data['Distance_km'] = data['Distance_km'].astype(float)
data['Fare'] = data['Fare'].astype(float)
data['Cost_per_passenger'] = data['Cost_per_passenger'].astype(float)
data['Passengers'] = data['Passengers'].fillna(0).astype(int)

data
```

```python
# Save cleaned data
data.to_csv('metro_data.csv', index=False)
```
